## Pandas Built-In Data Visualization
Pandas has built-in capabilities for data visualization! It's built-off of matplotlib, but it baked into pandas for easier usage!

```python
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
df1 = pd.read_csv('df1',index_col=0)
df2 = pd.read_csv('df2')
```

```python
df1.head()         # This is a time-series data beacuse of the dates....
```

```python
df2.head()
```

## Creating Plots With Pandas
There are several plot types built-in to pandas, most of them statistical plots by nature:

    * df.plot.area
    * df.plot.barh
    * df.plot.density
    * df.plot.hist
    * df.plot.line
    * df.plot.scatter
    * df.plot.bar
    * df.plot.box
    * df.plot.hexbin
    * df.plot.kde
    * df.plot.pie
    
You can also just call df.plot(kind='hist') or replace that kind argument with any of the key terms shown in the list above (e.g. 'box','area', etc..)



### Histogram

```python
# Ex 1:- Getting a histogram for all of the values in column A for df1...

df1["A"].hist()
```

```python
df1["A"].hist(bins=30, ec="black")
```

```python
# 2nd method of calling the plot in Ex 1 is:

df1["A"].plot(kind="hist", bins=30, ec="black")
```

```python
# 3rd method of calling the plot in Ex 1 is:

df1["A"].plot.hist(bins=30, ec="black")
```

### Area Plot

```python
df2.plot.area()
```

```python
# Adding keyword arg - "alpha", for more transparency....

df2.plot.area(alpha=0.3)
```

### Barplot

```python
# Each bar represents each column: a,b,c,d from the rows: 0,1,2,3,4.... in the DataFrame

df2.plot.bar()
```

```python
# Adding a stack in the barplot

df2.plot.bar(stacked=True)
```

### Line PLot

```python
df1["A"].plot(kind="line", x=df1.index, y='B')
```

```python
# Since the line plot above is built off matplotlib, lets add some matplotlib keyword args...

df1["A"].plot(kind="line", x=df1.index, y='B', figsize=(10,3))
```

```python
# Making the lines skinner, we use the "linewidth" argument..

df1["A"].plot(kind="line", x=df1.index, y='B', figsize=(10,3), lw=0.8)
```

### Scatter Plot

```python
df1.plot.scatter(x="A", y="C")
```

```python
df1.plot.scatter(x="A", y="C", c="B")
```

```python
df1.plot.scatter(x="A", y="D", c="C", cmap="magma")
```

```python
# When Passing the size ("s") argument, you must add the DataFrame's column..

df1.plot.scatter(x="A", y="C", s=df1["B"])
```

```python
# The plot above is too small/tiny, so we multply by a value....
# Here we can see the points "A" versus "C" and the size of the data points indicates how large their "C" value was...

df1.plot.scatter(x="A", y="C", s=df1["B"]*50)
```

### BoxPlot

```python
df2.plot.box() # Can also pass a by= argument for groupby
```

### Hexagonal Bin Plot
This type of plot is useful for Bivariate Data and alternative to scatterplot.

```python
df = pd.DataFrame(np.random.randn(1000, 2), columns=['a', 'b'])
df.plot.hexbin(x='a',y='b')
```

```python
# To enlarge the size of the hexagons, we use the "gridsize" argument..

df = pd.DataFrame(np.random.randn(1000, 2), columns=['a', 'b'])
df.plot.hexbin(x='a',y='b',gridsize=22)
```

```python
# To change the colors of the Hexagon, we use color mapping arg - "cmap"

df = pd.DataFrame(np.random.randn(1000, 2), columns=['a', 'b'])
df.plot.hexbin(x='a',y='b',gridsize=25,cmap='magma')
```

## Kernel Density Estimation (KDE) plot

```python
df2['a'].plot.kde()
```

```python
df2['a'].plot.density()
```

```python
# The kde plot can be done for the DataFrame as well and not just for a column...

df2.plot.kde()
```

## Style Sheets
Matplotlib has style sheets you can use to make your plots look a little nicer. These style sheets include plot_bmh,plot_fivethirtyeight,plot_ggplot and more. They basically create a set of style rules that your plots follow. I use them because they make all your plots have the same look and feel more professional.

```python
df1['A'].hist(ec="black")
```

calling the styles ggplot

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')
df1['A'].hist(ec="purple")
```

calling the styles bmh

```python
plt.style.use('bmh')
df1['A'].hist(ec="black")
```

calling the styles dark_background

```python
plt.style.use('dark_background')
df1['A'].hist(ec="green")
```

calling the styles fivethirtyeight

```python
plt.style.use('fivethirtyeight')
df1['A'].hist(ec="orange")
```
